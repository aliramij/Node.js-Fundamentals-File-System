```python
# Question 4: Write a script that accesses and displays the command-line arguments passed to it when executed.

'''
// args.js
const args = process.argv.slice(2);

console.log("Arguments:", args);
console.log("Count:", args.length);

args.forEach((arg, index) => {
  console.log(`  [${index}] ${arg}`);
});

//in terminal run it
node args.js hello world 42

//output
Arguments: [ 'hello', 'world', '42' ]
Count: 3
  [0] hello
  [1] world
  [2] 42

'''
```

```python
# Question 5: Using the fs/promises module, write a script to asynchronously read the content of a file named data.txt and log it to the console.

'''
// read-file.js
const fs = require('fs/promises');

async function readFile() {
  const content = await fs.readFile('data.txt', 'utf8');
  console.log(content);
}

readFile();

//in terminal run it
node read-file.js
//output
Hello from data.txt!
'''
```

```python
# Question 6: Using the fs/promises module, write a script to create a new file named log.txt and write the string "Task completed." into it.

'''
// write-file.js
const fs = require('fs/promises');

async function writeFile() {
  await fs.writeFile('log.txt', 'Task completed.', 'utf8');
  console.log('File created and written successfully!');
}

writeFile();

//in terminal run it
node write-file.js

//output
File created and written successfully!
'''
```

```python
# Question 8: Create a basic HTTP server using the http module that listens on port 3000 and responds to all requests with "Server is running."

'''
// server.js
const http = require('http');

const server = http.createServer((req, res) => {
  res.writeHead(200, { 'Content-Type': 'text/plain' });
  res.end('Server is running.');
});

server.listen(3000, () => {
  console.log('Server listening on http://localhost:3000');
});

//in terminal run it
node server.js
//output
Server listening on http://localhost:3000
'''
```

```python
# Question 9: Enhance the previous HTTP server to handle two different routes: / should respond with "Homepage", and /about should respond with "About Us".

'''
// server.js
const http = require('http');

const server = http.createServer((req, res) => {
  const { method, url } = req;
  console.log(`[${new Date().toISOString()}] ${method} ${url}`);

  if (url === '/') {
    res.writeHead(200, { 'Content-Type': 'text/plain' });
    res.end('Homepage');
  } else if (url === '/about') {
    res.writeHead(200, { 'Content-Type': 'text/plain' });
    res.end('About Us');
  } else {
    res.writeHead(404, { 'Content-Type': 'text/plain' });
    res.end('404 Not Found');
  }
});

server.listen(3000, () => {
  console.log('Server listening on http://localhost:3000');
});

//in terminal run it
node server.js
'''
```

```python
# Question 10: Write a script that efficiently reads a large file (large.txt) and outputs its content to the console using streams.

'''
// read-stream.js
const fs = require('fs');

const stream = fs.createReadStream('large.txt', { encoding: 'utf8' });

stream.on('data', chunk => process.stdout.write(chunk));
stream.on('end',  ()    => console.log('\n--- Stream complete ---'));
stream.on('error', err  => console.error('Stream error:', err.message));

//in terminal run it
node read-stream.js

'''
```
